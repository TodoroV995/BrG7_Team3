```python
def get_minimal_pos_tag(raw_tag: str ) -> str:
  if raw_tag.startswith("N"):
    return 'n'
  if raw_tag.startswith("V"):
    return 'v'
  if raw_tag.startswith("J"):
    return 'a'
  return None
```
